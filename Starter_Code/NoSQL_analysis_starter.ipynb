```python
# Import dependencies
from pymongo import MongoClient
from pprint import pprint
import pandas as pd
```

```python
# assign the collection to a variable
establishments = db['establishments']
```

### 1. Which establishments have a hygiene score equal to 20?

```python
pprint(establishments.find_one())
```

```python
# Find the establishments with a hygiene score of 20
query = {'scores.Hygiene': {'$eq': 20}}

# Use count_documents to display the number of documents in the result
print(f"Number of results: {db.establishments.count_documents(query)}")

# Display the first document in the results using pprint
pprint(establishments.find_one(query))
```

```python
# Convert the result to a Pandas DataFrame
results = establishments.find(query)
results_df = pd.DataFrame(results)
# Display the number of rows in the DataFrame
print(f"Number of rows: {len(results_df.index)}")
# Display the first 10 rows of the DataFrame
results_df.head(10)
```

### 2. Which establishments in London have a `RatingValue` greater than or equal to 4?

```python
# Find the establishments with London as the Local Authority and has a RatingValue greater than or equal to 4.
london_query = {'LocalAuthorityName': {'$regex': 'London'}, 'RatingValue': {'$gte': 4}}

# Use count_documents to display the number of documents in the result
print(f"Number of results: {db.establishments.count_documents(london_query)}")
# Display the first document in the results using pprint
db.establishments.find_one(london_query)
```

```python
# Convert the result to a Pandas DataFrame
london_results = establishments.find(london_query)
london_results_df = pd.DataFrame(london_results)

# Display the number of rows in the DataFrame
print(f"Number of rows: {len(london_results_df.index)}")

# Display the first 10 rows of the DataFrame
london_results_df.head(10)
```

### 4. How many establishments in each Local Authority area have a hygiene score of 0?

```python
# Create a pipeline that: 
# 1. Matches establishments with a hygiene score of 0
# 2. Groups the matches by Local Authority
# 3. Sorts the matches from highest to lowest
match_query = {'$match': {'scores.Hygiene': {'$eq': 0}}}
group_query = {'$group': {'_id': 'LocalAuthorityName'}}
sort_values = {'$sort': {'count': -1}}

pipeline = [match_query, group_query, sort_values]

# Print the number of documents in the result
match_results = list(establishments.aggregate(pipeline))
print(f"Number of documents in result: {len(match_results)}")

# Print the first 10 results
print(f"Number of docs in match_query: {db.establishments.count_documents(match_query)}")
```

```python
# Convert the result to a Pandas DataFrame
match_results_df = pd.DataFrame(match_results)

# Display the number of rows in the DataFrame
print(f"Number of rows: {len(match_results_df.index)}")

# Display the first 10 rows of the DataFrame
match_results_df.head(10)
```
